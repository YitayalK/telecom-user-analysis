```python
# Count missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Total number of values in each column
total_values = df.count() + missing_values

# Calculate percentage of missing values
percentage_missing = (missing_values / total_values) * 100

print("Percentage of Missing Values:\n", percentage_missing)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
#Identify Outliers
from scipy import stats
import numpy as np

z_scores = stats.zscore(df.select_dtypes(include=['float64', 'int64']))
abs_z_scores = np.abs(z_scores)
outliers = (abs_z_scores > 3).any(axis=1)   # Identify rows with any outlier
print(df[outliers].sum())     # Display outlier rows
```

Handset and Manufacturer Analysis

```python
#Identify the top 10 handsets used by customers
top_10_handsets = df['Handset Type'].value_counts().head(10)
print(top_10_handsets)
```

```python
#Identify the top 3 handset manufacturers
top_3_manufacturers = df['Handset Manufacturer'].value_counts().head(3)
print(top_3_manufacturers)
```

```python
#Identify the top 5 handsets per top 3 manufacturers
for manufacturer in top_3_manufacturers.index:
    top_handsets = df[df['Handset Manufacturer'] == manufacturer]['Handset Type'].value_counts().head(5)
    print(f"Top handsets for {manufacturer}:")
    print(top_handsets)
```

```python
#Analyze User Behavior
user_behavior = df.groupby('MSISDN/Number').agg({
    'Dur. (ms)': 'sum',
    'Total DL (Bytes)': 'sum',
    'Total UL (Bytes)': 'sum',
    'Social Media DL (Bytes)': 'sum',
    'Google DL (Bytes)': 'sum',
    # Add other relevant columns...
}).reset_index()
```

```python
#Segment users into deciles based on total session duration
user_behavior['Decile'] = pd.qcut(user_behavior['Dur. (ms)'], q=10, labels=False)
```

```python
user_behavior.describe()
```

```python
#Visualize distributions
sns.histplot(user_behavior['Total DL (Bytes)'], bins=30, kde=True)
plt.show()
```

```python
#relationships between applications and total data
sns.scatterplot(data=user_behavior, x='Social Media DL (Bytes)', y='Total DL (Bytes)')
plt.show()
```

```python
#Compute a correlation matrix for application usage
correlation_matrix = user_behavior.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
#PCA to reduce dimensionality
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
reduced_data = pca.fit_transform(user_behavior[["Total DL (Bytes)", "Total UL (Bytes)", "Dur. (ms)"]])
print(pca.explained_variance_ratio_)
```
